```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/dataset.csv")

df.columns = df.columns.str.lower()

print(df.shape)
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['type'].fillna('Unknown').value_counts().plot(kind='bar', figsize=(8,4))
plt.title('Distribusi type anime')
plt.ylabel('Jumlah')
plt.tight_layout()
plt.show()
```

```python
df['status'].fillna('Unknown').value_counts().plot(kind='bar', figsize=(8,4))
plt.title('Distribusi status anime')
plt.tight_layout()
plt.show()
```

```python
df['genres'] = df['genres'].fillna('Unknown')

genre_counts = (
  df['genres']
  .str.split(', ')
  .explode()
  .value_counts()
  .head(20)
)

plt.figure(figsize=(10,5))
genre_counts.plot(kind='bar')
plt.title('20 Genre anime terbanyak')
plt.tight_layout()
plt.show()
```
